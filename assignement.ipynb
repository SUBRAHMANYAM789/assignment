```python
#29 Solve the system of linear equations 2x + 3y = 5 and 4x + 6y = 10 using NumPy
arr1=np.array([[2,3],[4,6]])

arr2=np.array([5,10])

rank_1 = np.linalg.matrix_rank(arr1)

rank_12 = np.linalg.matrix_rank(np.column_stack((arr1, arr2)))

if rank_1 == rank_12 == 2:
    # Unique solution
    solution = np.linalg.solve(arr1, arr2)
    print("Unique solution:", solution)
elif rank_1 == rank_12 < 2:
    print("Infinite solutions (dependent system)")
else:
    print("No solution (inconsistent system)")
```

```python
#30

A = np.array([[3, 2, 2],
              [2, 3, -2],
              [2, -2, 3]])

#Perform SVD
U, S, VT = np.linalg.svd(A)

print("U:\n", U)
print("S (singular values):\n", S)
print("VT:\n", VT)

# Reconstruct the original matrix
# Convert S (which is a 1D array of singular values) into a 3x3 diagonal matrix
S_matrix = np.zeros((3, 3))
np.fill_diagonal(S_matrix, S)

# A = U * S * VT
A_reconstructed = U @ S_matrix @ VT

print("Reconstructed A:\n", A_reconstructed)
```

#3. Merging and Joining Datasets


```python
#Merge two DataFrames based on a common key and fill any missing values in the resulting DataFrame.

df1 = pd.DataFrame({
    'id': [1, 2, 3],
    'name': ['Alice', 'Bob', 'Charlie']
})

df2 = pd.DataFrame({
    'id': [2, 3, 4],
    'score': [88, None, 75]
})

merged_df = pd.merge(df1, df2, on='id', how='outer')

#  Fill missing values (e.g., with default text/number)
merged_df.fillna({'name': 'Unknown', 'score': 0}, inplace=True)

print(merged_df)
```

```python
#Perform a left join on two DataFrames with different keys and handle missing data in the result.
employees = pd.DataFrame({
    'emp_id': [1, 2, 3],
    'name': ['Alice', 'Bob', 'Charlie']
})

salaries = pd.DataFrame({
    'employee_id': [2, 3, 4],
    'salary': [50000, None, 60000]
})

merged_df = pd.merge(employees, salaries, how='left', left_on='emp_id', right_on='employee_id')

merged_df.fillna({'salary': salaries.salary.mean(),"employee_id":employees.emp_id}, inplace=True)

print(merged_df)
```

```python
# Concatenate two DataFrames along the columns and handle any duplicate column names.
df1 = pd.DataFrame({
    'id': [1, 2, 3],
    'value': ['A', 'B', 'C']
})

df2 = pd.DataFrame({
    'value': [10, 20, 30],
    'score': [90, 85, 95]
})

df2=df2.rename(columns={"value":"values"})
df=pd.concat([df1,df2])

print(df)
```

#6. Broadcasting and Vectorized Operations


```python
#Broadcast a NumPy array across a DataFrame column to perform a vectorized operation.
df = pd.DataFrame({
    'Product': ['A', 'B', 'C', 'D'],
    'Price': [100, 200, 300, 400]
})

# NumPy array representing a discount (e.g., 10%, 20%, 30%, 40%)
discounts = np.array([0.10, 0.20, 0.30, 0.40])

# Vectorized operation: apply discount using broadcasting
discounted_price = df['Price'] * (1 - discounts)

# Assign the result to a new column
df['Discounted_Price'] = discounted_price

print(df)
```

```python
#Create a new column in a DataFrame that results from a vectorized operation on multiple columns using NumPy.
df = pd.DataFrame({
    'Quantity': [2, 5, 3, 4],
    'Unit_Price': [100, 200, 150, 300],
    'Tax_Rate': [0.05, 0.1, 0.08, 0.12]  # e.g., 5%, 10%, etc.
})

# Vectorized calculation of total cost with tax
# Formula: (Quantity * Unit_Price) * (1 + Tax_Rate)
total_cost = np.array(df['Quantity']) * np.array(df['Unit_Price']) * (1 + np.array(df['Tax_Rate']))

# Add result as new column
df['Total_Cost'] = total_cost

print(df)
```

```python
# Demonstrate broadcasting by subtracting the mean of each row from the row's elements in a DataFrame.

df = pd.DataFrame({
    'A': [10, 20, 30],
    'B': [20, 30, 40],
    'C': [30, 40, 50]
})

print("Original DataFrame:")
print(df)

# Subtract row mean using broadcasting
row_means = df.mean(axis=1)
df_centered = df.sub(row_means, axis=0)

print("\nDataFrame after subtracting row mean:")
print(df_centered)
```

```python
# Create a 3D array using NumPy and convert it into a MultiIndex DataFrame, then perform a group operation.

# Shape: (2, 3, 4) => 2 blocks, 3 rows, 4 columns
array_3d = np.random.randint(1, 10, size=(2, 3, 4))

# We'll reshape it to 2D first: shape (2*3, 4)
reshaped_array = array_3d.reshape(-1, array_3d.shape[2])

# Create MultiIndex: (block_id, row_id)
block_ids = np.repeat(range(array_3d.shape[0]), array_3d.shape[1])
row_ids = list(range(array_3d.shape[1])) * array_3d.shape[0]
index = pd.MultiIndex.from_arrays([block_ids, row_ids], names=['block', 'row'])

# Create DataFrame
df = pd.DataFrame(reshaped_array, index=index, columns=[f'col_{i}' for i in range(array_3d.shape[2])])

print("Original MultiIndex DataFrame:")
print(df)

# Step 3: Perform a group operation
# Example: Group by 'block' and calculate the mean of each column
grouped_mean = df.groupby(level='block').mean()

print("\nGrouped by 'block' and averaged:")
print(grouped_mean)
```

#12. Time Series Analysis

```python
#Convert a column in a DataFrame to a time series using Pandas and perform a NumPy-based operation on the time series data.

dates = pd.date_range(start='2023-01-01', periods=10, freq='D')
values = np.random.randint(50, 100, size=10)

df = pd.DataFrame({
    'Date': dates,
    'Value': values
})

# Set the 'Date' column as the index to create a time series
df.set_index('Date', inplace=True)

print("Time Series DataFrame:")
print(df)

# Example: Calculate a rolling z-score for the 'Value' column

# Calculate mean and std using NumPy
mean = np.mean(df['Value'])
std = np.std(df['Value'])

# Compute z-score
df['Z_Score'] = (df['Value'] - mean) / std

print(df)
```

```python
# Calculate the moving average of a time series in a DataFrame using a combination of Pandas and NumPy functions.
dates = pd.date_range(start='2023-01-01', periods=15, freq='D')
values = np.random.randint(10, 100, size=15)

df = pd.DataFrame({
    'Date': dates,
    'Value': values
})
df.set_index('Date', inplace=True)

print("Original Time Series:")
print(df)

df['Moving_Avg_Pandas'] = df['Value'].rolling(window=3).mean()

# We'll use convolution to simulate moving average
weights = np.ones(3) / 3
moving_avg_numpy = np.convolve(df['Value'], weights, mode='valid')

# To align with the index (pad with NaNs to match shape)
df['Moving_Avg_Numpy'] = np.nan
df.iloc[2:, df.columns.get_loc('Moving_Avg_Numpy')] = moving_avg_numpy

print("\nTime Series with Moving Averages:")
print(df)
```

```python
#Use pd.to_datetime to convert a string column to datetime format and use NumPy to compute the time difference between rows.

df = pd.DataFrame({
    'Date': ['2023-01-01', '2023-01-03', '2023-01-05', '2023-01-10']
})

df['Date'] = pd.to_datetime(df['Date'])

print("DataFrame with datetime:")
print(df)

# We'll shift the datetime column to get previous date
date_array = df['Date'].values

# compute the difference in seconds
time_diff = np.diff(date_array)

df['Days_Diff'] = pd.Series([np.nan] + [td / np.timedelta64(1, 'D') for td in time_diff])

print("\nDataFrame with time difference between rows (in days):")
print(df)
```

```python
# 10. Implement a function using NumPy that returns the inverse of a given matrix.
import numpy as np

def invert_matrix(matrix):
    try:
        inv_matrix = np.linalg.inv(matrix)
        return inv_matrix
    except np.linalg.LinAlgError as e:
        print("Matrix inversion error:", e)
        return None

# Example usage
A = np.array([[1, 2],
              [3, 4]])

inverse_A = invert_matrix(A)
print("Original Matrix:\n", A)
print("Inverse Matrix:\n", inverse_A)
```

```python
# 30. Perform feature engineering on a dataset by creating new features and then visualize their importance using a bar chart.

df = pd.DataFrame({
    'height_cm': [150, 160, 170, 180, 190],
    'weight_kg': [50, 60, 65, 80, 85],
    'age': [22, 25, 30, 35, 40],
    'target': [0, 0, 1, 1, 1]  # some binary target
})


df['bmi'] = df['weight_kg'] / ((df['height_cm'] / 100) ** 2)
df['is_adult'] = (df['age'] >= 18).astype(int)
df['age_squared'] = df['age'] ** 2

print(df)
```

```python
from sklearn.ensemble import RandomForestClassifier
import matplotlib.pyplot as plt

# Define features and target
X = df.drop(columns=['target'])  # features
y = df['target']                 # target

# Train model
model = RandomForestClassifier(random_state=42)
model.fit(X, y)

# Get feature importances
importances = model.feature_importances_
feature_names = X.columns

# Create DataFrame for plotting
importance_df = pd.DataFrame({
    'Feature': feature_names,
    'Importance': importances
}).sort_values(by='Importance', ascending=False)

# Plot
plt.figure(figsize=(8, 5))
plt.barh(importance_df['Feature'], importance_df['Importance'], color='skyblue')
plt.xlabel('Importance Score')
plt.title('Feature Importance')
plt.gca().invert_yaxis()  # Highest at top
plt.tight_layout()
plt.show()
```
